#### LMS Estimation of $\Theta$ based on X

Let's assume a simple linear model where we want to estimate a variable $\Theta$ based on an observation $X$. Our true underlying relationship is $\Theta = aX + b + \text{noise}$. We restrict our estimator to be linear: $\hat{\Theta} = \alpha X + \beta$.

The goal of LLMS is to find $\alpha$ and $\beta$ that minimize the Mean Squared Error (MSE): $\text{MSE} = E[(\Theta - \hat{\Theta})^2] = E[(\Theta - (\alpha X + \beta))^2]$

```python
# True parameters
true_a, true_b = 2.5, 5.0 # Slope, Intercept
sigma_noise = 2.0 # Standard deviation of noise

# Generate X values and noise
np.random.seed(42) # for reproducibility
X = np.random.uniform(0, 10, 100) # 100 data points between 0 and 10
noise = np.random.normal(0, sigma_noise, X.shape)

# Generate Theta (the true underlying values)
Theta = true_a * X + true_b + noise
```

#### LLMS Estimator
**Derivation of LLMS Estimator Parameters**

To minimize the MSE, we take the partial derivatives with respect to $\alpha$ and $\beta$ and set them to zero.

$\frac{\partial \text{MSE}}{\partial \beta} = 2 E[(\Theta - \alpha X - \beta)(-1)] = 0$
$E[\Theta] - \alpha E[X] - \beta = 0$

From this, we get $\beta = E[\Theta] - \alpha E[X]$ (Equation 1)

$\frac{\partial \text{MSE}}{\partial \alpha} = 2 E[(\Theta - \alpha X - \beta)(-X)] = 0$
$E[X\Theta] - \alpha E[X^2] - \beta E[X] = 0$

Substitute Equation 1 into the second derivative equation:
$E[X\Theta] - \alpha E[X^2] - (E[\Theta] - \alpha E[X]) E[X] = 0$.
$E[X\Theta] - \alpha E[X^2] - E[\Theta]E[X] + \alpha (E[X])^2 = 0$
$\alpha (E[X]^2 - E[X^2]) = E[\Theta]E[X] - E[X\Theta]$
$\alpha = \frac{E[X\Theta] - E[X]E[\Theta]}{E[X^2] - E[X]^2}$

Recognizing that $E[X\Theta] - E[X]E[\Theta] = \text{Cov}(X, \Theta)$ and $E[X^2] - E[X]^2 = \text{Var}(X)$, we get $\alpha = \frac{\text{Cov}(X, \Theta)}{\text{Var}(X)}$

And from first Equation, we have $\beta = E[\Theta] - \alpha E[X]$

These are the well-known formulas for the coefficients of the simple linear regression line (which is the LLMS estimator in this case).

```python
# Implement the LLMS Estimator

# Calculate the expectations (using sample means as estimates for expectations)
EX, ETheta = np.mean(X), np.mean(Theta)
EX2, EXTheta = np.mean(X**2), np.mean(X * Theta)

# Calculate alpha and beta using the derived formulas
alpha_llms = (EXTheta - EX * ETheta) / (EX2 - EX**2)
beta_llms = ETheta - alpha_llms * EX

# Generate predictions using the LLMS estimator
Theta_hat = alpha_llms * X + beta_llms

# Plot the estimated relationship along with the true and observed data
```

**LLMS Performance Analysis**

To evaluate the performance of our LLMS estimator, we can calculate metrics such as the Mean Squared Error (MSE) and the coefficient of determination (R-squared). These metrics will help us understand how well our estimated model explains the variance in $\Theta$.

**Mean Squared Error** $\text{MSE} = \frac{1}{N} \sum_{i=1}^{N} (\Theta_i - \hat{\Theta}_i)^2$

**R-squared** $R^2 = 1 - \frac{\sum (\Theta_i - \hat{\Theta}_i)^2}{\sum (\Theta_i - \bar{\Theta})^2} = 1 - \frac{\text{SSE}}{\text{SST}}$
where
* SSE is the Sum of Squared Errors (residuals)
* SST is the Total Sum of Squares
* $\bar{\Theta}$ is the mean of the true $\Theta$ values.

```python
# Mean Squared Error (MSE)
mse = np.mean((Theta - Theta_hat)**2)

# R-squared
ss_total = np.sum((Theta - np.mean(Theta))**2) # Total sum of squares
ss_residual = np.sum((Theta - Theta_hat)**2) # Sum of squares of residuals
r_squared = 1 - (ss_residual / ss_total)

# Compare MSE to the variance of the noise (theoretical lower bound for unbiased estimator)

# Let's check the MSE if we just used the true parameters (without noise in X)
# This helps understand the impact of the noise itself
theta_true_no_noise = true_a * X + true_b
mse_true_model = np.mean((Theta - theta_true_no_noise)**2)
```

### LLMS Estimation of U based on X, where X = U + W

Implements the LLMS estimator for a scenario where we observe $X = U + W$, and we want to estimate U.

**Given properties** $X = U + W$, U and W are independent, E[W] = 0

**Derived LLMS estimator of U** (denoted as $U_hat = a*X + b$)
*   $a = Var(U) / (Var(U) + Var(W))$, $b = E[U] * Var(W) / (Var(U) + Var(W))$

```python
# Define True Parameters for U and W

# Properties of U: E[U], Std Dev of U, Var(U)
mu_u, sigma_u = 10.0, 3.0
sigma_u_sq = sigma_u**2

# Properties of W (noise)
sigma_w = 2.0    # Std Dev of W
sigma_w_sq = sigma_w**2 # Var(W)
# E[W] is implicitly 0 for standard noise models

# Number of samples
n_samples = 200
np.random.seed(44) # for reproducibility

# Generate U values
U = np.random.normal(loc=mu_u, scale=sigma_u, size=n_samples)
# Generate W (noise) values
W = np.random.normal(loc=0, scale=sigma_w, size=n_samples)
# Generate X (observation)
X = U + W
```

```python
# Implement the LLMS Estimator for U based on X

# Calculate coefficients 'a' and 'b' using the derived formulas
# Note: We use the true population parameters E[U], Var(U), Var(W) for 'a' &'b'
# as the problem statement implies these are known.

a_llms_uw = sigma_u_sq / (sigma_u_sq + sigma_w_sq)
b_llms_uw = mu_u * sigma_w_sq / (sigma_u_sq + sigma_w_sq)

# Apply the estimator to X to get U_hat
U_hat = a_llms_uw * X + b_llms_uw

# Evaluate Performance
# Mean Squared Error (MSE) of U_hat in estimating U
mse_uw = np.mean((U - U_hat)**2)

# Calculate R-squared for this estimation
ss_total_uw = np.sum((U - np.mean(U))**2)
ss_residual_uw = np.sum((U - U_hat)**2)
r_squared_uw = 1 - (ss_residual_uw / ss_total_uw)

# Theoretical minimum MSE for this problem (related to Var(U|X))
# For independent U and W, U|X is normal with mean aX+b and variance Var(U) - Cov(U,X)^2/Var(X) = Var(U) - a*Cov(U,X) = Var(U) - a*Var(U) = Var(U)*(1-a) = Var(U) * Var(W) / (Var(U) + Var(W)) = sigma_u_sq * sigma_w_sq / (sigma_u_sq + sigma_w_sq)
# This is the conditional variance of U given X, which is the theoretical minimum MSE for an unbiased estimator.
# The LLMS estimator achieves this minimum variance.
min_mse_theoretical = (sigma_u_sq * sigma_w_sq) / (sigma_u_sq + sigma_w_sq)
```

### LLMS Estimation with Multiple Independent Variables**

Consider a scenario where $\Theta$ depends on multiple independent variables, say $X_1, X_2, \dots, X_p$. Our true underlying relationship is $\Theta = a_1 X_1 + a_2 X_2 + \dots + a_p X_p + b + \text{noise}$. Our linear estimator will be
$\hat{\Theta} = \alpha_1 X_1 + \alpha_2 X_2 + \dots + \alpha_p X_p + \beta$.

In matrix form, this can be written as $\hat{\Theta} = \mathbf{X}_e \mathbf{w}$, where $\mathbf{X}_e$ is the extended design matrix (including a column of ones for the intercept), and $\mathbf{w}$ is the vector of coefficients (including the intercept $\beta$).

The goal remains to minimize the Mean Squared Error $E[(\Theta - \hat{\Theta})^2]$.

```python
# Synthetic Data for Multiple Variables
# True parameters for multiple features
true_weights = np.array([2.5, -1.0, 3.0]) # Coefficients for X1, X2, X3
true_intercept = 5.0
sigma_noise_multi = 2.0 # Standard deviation of noise
# Number of data points and features
n_samples, n_features = 100, len(true_weights)

# Generate multiple X values
np.random.seed(43) # for reproducibility
X_multi = np.random.uniform(0, 10, (n_samples, n_features)) # 100 data points, 3 features

# Generate noise
noise_multi = np.random.normal(0, sigma_noise_multi, n_samples)

# Generate Theta (the true underlying values)
Theta_multi = X_multi @ true_weights + true_intercept + noise_multi
```

**Derivation of LLMS Estimator Parameters for Multiple Variables (Matrix Form)**

For multiple linear regression, the coefficients $\mathbf{w}$ that minimize the MSE can be found using the normal equation. This equation is derived similarly by setting the partial derivatives of the MSE with respect to each coefficient to zero.

The solution for the optimal weight vector $\mathbf{w}$ is given by $\mathbf{w} = (\mathbf{X}_e^T \mathbf{X}_e)^{-1} \mathbf{X}_e^T \mathbf{\Theta}$,

where
*   $\mathbf{w}$ is the column vector of estimated coefficients ($\alpha_1, \dots, \alpha_p, \beta)^T$
*   $\mathbf{X}_e$ is the design matrix, which includes a column of ones for the intercept. If $\mathbf{X}$ is the matrix of independent variables (shape $N \times P$), then $\mathbf{X}_e$ will have shape $N \times (P+1)$.
*   $\mathbf{\Theta}$ is the column vector of observed dependent variables (shape $N \times 1$).

```python
# Implement the Multiple LLMS Estimator

# Add a column of ones to X_multi for the intercept term
X_e = np.hstack((X_multi, np.ones((n_samples, 1))))

# Calculate the coefficients using the Normal Equation
# w = (X_e.T @ X_e)^-1 @ X_e.T @ Theta_multi
weights_llms_multi = np.linalg.inv(X_e.T @ X_e) @ X_e.T @ Theta_multi

# Extract estimated alpha (weights for features) and beta (intercept)
alpha_llms_multi = weights_llms_multi[:-1]
beta_llms_multi = weights_llms_multi[-1]

# Generate predictions using the multiple LLMS estimator
Theta_hat_multi = X_e @ weights_llms_multi

# Evaluate Performance
# Mean Squared Error (MSE)
mse_multi = np.mean((Theta_multi - Theta_hat_multi)**2)

# R-squared
ss_total_multi = np.sum((Theta_multi - np.mean(Theta_multi))**2)
ss_residual_multi = np.sum((Theta_multi - Theta_hat_multi)**2)
r_squared_multi = 1 - (ss_residual_multi / ss_total_multi)
```

### LLMS Estimation with Random Sums

Let's analyze the problem where we want to estimate a random variable $N$ (number of terms) based on a random sum $S_A = \sum_{i=1}^{N} A_i$. Our estimator will be of the form $\hat{N} = c_1 S_A + c_2$.

**Given Properties:**
*   $N$: Random variable, $E[N] = m$, $Var(N) = v$.
*   $A_i$: Sequence of i.i.d. r.v.s, independent of $N$. $E[A_i] = 1$, $Var(A_i) = 1$.
*   $B_i$: Sequence of i.i.d. r.v.s, independent of $N$ and $A_i$. $E[B_i] = 1$, $Var(B_i) = 1$.

We need to find $c_1$ and $c_2$ that minimize $E[(N - \hat{N})^2]$.

The general formulas for LLMS coefficients are:
$c_1 = \frac{Cov(S_A, N)}{Var(S_A)}$, $c_2 = E[N] - c_1 E[S_A]$

**Derivation of Required Expected Values and Variances**

**1. $E[AB]$**: Since $A_i$ and $B_j$ are independent for all $i,j$ (the problem states $B_i$ are independent of $A_i$), then:
$E[A_i B_j] = E[A_i] E[B_j] = 1 \cdot 1 = 1$
So, generally, $E[AB] = 1$.

**2. $E[NA_i]$**: Since $N$ and $A_i$ are independent:
$E[N A_i] = E[N] E[A_i] = m \cdot 1 = m$

**3. $E[S_A]$** (where $S_A = \sum_{i=1}^{N} A_i$)
Using Wald's Identity (or conditioning on $N$):
$E[S_A] = E[E[S_A | N]] = E[N \cdot E[A_1]] = E[N \cdot 1] = E[N] = m$.

**4. $Var(S_A)$**
Using the Law of Total Variance: $Var(S_A) = E[Var(S_A | N)] + Var(E[S_A | N])$
*   $Var(S_A | N) = N \cdot Var(A_1) = N \cdot 1 = N$
*   So, $E[Var(S_A | N)] = E[N] = m$.
*   $E[S_A | N] = N \cdot E[A_1] = N \cdot 1 = N$
*   So, $Var(E[S_A | N]) = Var(N) = v$.

Therefore, $Var(S_A) = m + v$.

**5. $Cov(S_A, N)$**
$Cov(S_A, N) = E[S_A N] - E[S_A] E[N]$
We know $E[S_A] = m$ and $E[N] = m$. So, $Cov(S_A, N) = E[S_A N] - m^2$.

To find $E[S_A N]$:
$E[S_A N] = E[E[S_A N | N]] = E[N \cdot E[S_A | N]] = E[N \cdot N] = E[N^2]$
From the definition of variance, $Var(N) = E[N^2] - (E[N])^2$, so $E[N^2] = Var(N) + (E[N])^2 = v + m^2$.

Therefore, $Cov(S_A, N) = (v + m^2) - m^2 = v$.

**Derivations for Random Sums ($S_A = \sum_{i=1}^{N} A_i$, $S_B = \sum_{j=1}^{N} B_j$)**

**1. $E[S_A S_B]$**

Using the law of total expectation:
$E[S_A S_B] = E[E[S_A S_B | N]]$

Given $N=n$, we have $S_A = \sum_{i=1}^{n} A_i$ and $S_B = \sum_{j=1}^{n} B_j$. Since $A_i$ and $B_j$ are independent of $N$ and of each other
- $E[S_A S_B | N=n] = E\left[\left(\sum_{i=1}^{n} A_i\right) \left(\sum_{j=1}^{n} B_j\right) \bigg| N=n\right] = E\left[\sum_{i=1}^{n} \sum_{j=1}^{n} A_i B_j \bigg| N=n\right] = \sum_{i=1}^{n} \sum_{j=1}^{n} E[A_i B_j | N=n]$

Since $A_i$ and $B_j$ are independent
- $E[A_i B_j | N=n] = E[A_i B_j] = E[A_i] E[B_j] = 1 \cdot 1 = 1$. So, $E[S_A S_B | N=n] = \sum_{i=1}^{n} \sum_{j=1}^{n} 1 = n \cdot n = n^2$.

Therefore, $E[S_A S_B] = E[N^2]$. From the definition of variance, $Var(N) = E[N^2] - (E[N])^2$,
- $E[N^2] = Var(N) + (E[N])^2 = v + m^2$.

**Result for $E[S_A S_B]$:** $E[S_A S_B] = v + m^2$

**2. $E[S_A N]$**

This expectation was derived in the previous section (6. LLMS Estimation with Random Sums) when calculating $Cov(S_A, N)$. We found that $E[S_A N] = E[E[S_A N | N]] = E[N \cdot E[S_A | N]] = E[N \cdot N] = E[N^2]$. And as shown above, $E[N^2] = Var(N) + (E[N])^2 = v + m^2$.

**Result for $E[S_A N]$** $E[S_A N] = v + m^2$

**Derivation of LLMS Estimator Coefficients $c_1$ and $c_2$**

Now, substitute the derived components into the formulas for $c_1$ and $c_2$:

$c_1 = \frac{Cov(S_A, N)}{Var(S_A)} = \frac{v}{m + v}$

$c_2 = E[N] - c_1 E[S_A] = m - \left(\frac{v}{m + v}\right) m = m \left(1 - \frac{v}{m + v}\right) = m \left(\frac{m + v - v}{m + v}\right) = m \left(\frac{m}{m + v}\right) = \frac{m^2}{m + v}$

**The LLMS Estimator of N given $S_A$ is:**
- $\hat{N} = \left(\frac{v}{m + v}\right) S_A + \left(\frac{m^2}{m + v}\right)$

```python
# Define True Parameters for N, A_i, B_i
# Properties of N
m, v = 10.0, 5.0 # E[N] Var(N)

# Properties of A_i (mean 1, var 1)
sigma_A, sigma_B = 1.0, 1.0

# Number of trials (simulations)
num_trials = 1000
np.random.seed(46) # for reproducibility

# Generate Synthetic Data
# Generate N values based on mean m and variance v
# Using a Normal dist. and ensuring positivity and integer values for count
# Note: A true count RV like Poisson or Negative Binomial might be more accurate if specific distributions were assumed for N
N_actual_vals = np.random.normal(loc=m, scale=np.sqrt(v), size=num_trials)
N_vals = np.maximum(1, np.round(N_actual_vals)).astype(int)

# Generate A_i values for each N (random sums)
S_A_vals = np.zeros(num_trials)
actual_N_sums = []

for i in range(num_trials):
    current_N = N_vals[i]
    actual_N_sums.append(current_N) # Store the actual N for comparison
    A_i_samples = np.random.normal(loc=1, scale=sigma_A, size=current_N)
    S_A_vals[i] = np.sum(A_i_samples)
```

```python
# Implement the LLMS Estimator for N based on S_A
# Calculate coefficients c1 and c2 using the derived formulas
# We use the *true* population parameters m and v for the coefficients
c1_llms = v / (m + v)
c2_llms = (m**2) / (m + v)

# Apply the estimator to S_A to get N_hat
N_hat = c1_llms * S_A_vals + c2_llms

# Evaluate Performance
# Mean Squared Error (MSE) of N_hat in estimating N
mse_nsa = np.mean((N_vals - N_hat)**2)

# Calculate R-squared for this estimation
ss_total_nsa = np.sum((N_vals - np.mean(N_vals))**2)
ss_residual_nsa = np.sum((N_vals - N_hat)**2)
r_squared_nsa = 1 - (ss_residual_nsa / ss_total_nsa)
```

### Time Series Forecasting with LLMS

Now, let's consider a time series forecasting problem where we want to estimate a future value of a time-dependent variable, say $Y_t$, based on past observations or the time index itself. We generate synthetic time series data that exhibits a trend, seasonality, and random noise.

Our goal will be to apply LLMS principles to model this time series.

**Applying LLMS to Forecast the Time Series**

We build an LLMS model to forecast $Y$_$ts$ based on its time index. Since our synthetic data includes both a linear trend and seasonality, we can model this using a linear combination of the time index and a sinusoidal function of time.

Our estimator will look like
- $\hat{Y}_t = \beta_0 + \beta_1 \cdot \text{Time}_t + \beta_2 \cdot \sin(\text{Time}_t \cdot \text{seasonality_frequency})$

This is a form of multiple linear regression where Time and sin(Time * seasonality_frequency) are our independent variables.

```python
# Generate Synthetic Time Series Data
np.random.seed(45) # for reproducibility
n_steps = 100 # Number of time steps
# Time index
time = np.arange(n_steps)
# Linear Trend
trend = 0.5 * time + 10
# Seasonality (e.g., a sine wave, representing daily/weekly patterns)
seasonality_amplitude = 5
seasonality_frequency = 0.1 # Adjust for faster/slower cycles
seasonality = seasonality_amplitude * np.sin(time * seasonality_frequency)
# Random Noise
noise_std = 2.5
noise = np.random.normal(0, noise_std, n_steps)
# Combine components to get the synthetic time series Y
Y_ts = trend + seasonality + noise
# Create a Pandas DataFrame for easier handling of time series
df_ts = pd.DataFrame({'Time': time, 'Y_ts': Y_ts})
```

```python
# Prepare Features for Time Series LLMS
# Extract the Time variable from the DataFrame
time_feature = df_ts['Time'].values

# Create a seasonal feature based on the known seasonality_frequency
# We'll use the seasonality_frequency defined during data generation.
seasonal_feature = seasonality_amplitude * np.sin(time_feature * seasonality_frequency)

# Form the design matrix X for the LLMS model. It should include a
# column of ones for the intercept, the time feature, and the seasonal feature.
X_ts_features = np.vstack([np.ones(n_steps), time_feature, seasonal_feature]).T

# The target variable is Y_ts
Y_ts_target = df_ts['Y_ts'].values
```

```python
# Implement LLMS for Time Series Forecasting
# Calculate the coefficients (weights) using the Normal Equation
# w = (X_ts_features.T @ X_ts_features)^-1 @ X_ts_features.T @ Y_ts_target
weights_ts_llms = np.linalg.inv(X_ts_features.T @ X_ts_features) @ X_ts_features.T @ Y_ts_target

# Extract the estimated coefficients
intercept_ts_llms = weights_ts_llms[0]
beta1_ts_llms = weights_ts_llms[1]
beta2_ts_llms = weights_ts_llms[2]

# Generate predictions using the LLMS model
Y_ts_forecast = X_ts_features @ weights_ts_llms

# Evaluate Performance
mse_ts = np.mean((Y_ts_target - Y_ts_forecast)**2)
ss_total_ts = np.sum((Y_ts_target - np.mean(Y_ts_target))**2)
ss_residual_ts = np.sum((Y_ts_target - Y_ts_forecast)**2)
r_squared_ts = 1 - (ss_residual_ts / ss_total_ts)
```
