### Scenario 1:
Given $\Theta$ and $X$, two random variables not independent, and $E[\Theta | X = x]$ is a constant, define the Linear Least Mean Squares (LLMS) estimator in the form $aX + b$.

**Given:** $E[\Theta | X = x] = C$ for all $x$. **LLMS Estimator:** $\hat{\Theta} = aX + b$

**Derivation:**
1.  **$E[\Theta]$:** Since $E[\Theta | X] = C$, we have $E[\Theta] = E[E[\Theta | X]] = E[C] = C$.
2.  **$E[X\Theta]$:** $E[X\Theta] = E[E[X\Theta | X]] = E[X E[\Theta | X]] = E[X C] = C E[X]$.
3.  **$Cov(X, \Theta)$:** $Cov(X, \Theta) = E[X\Theta] - E[X]E[\Theta] = C E[X] - E[X]C = 0$.
4.  **Coefficient 'a':** $a = \frac{Cov(X, \Theta)}{Var(X)} = \frac{0}{Var(X)} = 0$ (assuming $Var(X) \neq 0$).
5.  **Coefficient 'b':** $b = E[\Theta] - a \cdot E[X] = C - 0 \cdot E[X] = C$.

**General Form of Estimator:** $\hat{\Theta} = C$

```python
import numpy as np

# Random seed for reproducibility
np.random.seed(42)

# Generate sample data for X
num_samples = 1000
X = np.random.rand(num_samples) * 10  # X values between 0 and 10

# Generate sample data for Theta such that E[Theta | X = x] = 3
# Theta = 3 + some random noise (independent of X)
# This ensures that the average of Theta for any X will be around 3
noise = np.random.randn(num_samples) # Standard normal noise
Theta = 3 + noise

# E[X]
E_X = np.mean(X)
# E[Theta]
E_Theta = np.mean(Theta)
# Cov(X, Theta) = E[X*Theta] - E[X]*E[Theta]
E_X_Theta = np.mean(X * Theta)
Cov_X_Theta = E_X_Theta - E_X * E_Theta
# Var(X) = E[X^2] - (E[X])^2
E_X_squared = np.mean(X**2)
Var_X = E_X_squared - (E_X)**2

# Calculate 'a' and 'b' for the LLS estimator

# Avoid division by zero if Var_X is extremely small (though unlikely with random data)
if Var_X > 1e-9:
    a = Cov_X_Theta / Var_X
else:
    a = 0 #If X has no variance, it's a constant, and a would be undefined or 0

b = E_Theta - a * E_X
```

### Scenario 2:

Given the random variables $X$, $\Theta$, and $W$ with the conditions
- $\Theta$ and $W$ are independent.
- $E[\Theta]$ and $E[W]$ are given.
- $Var(\Theta) = \sigma_{\Theta}^2$, $Var(W) = \sigma_W^2$
- $X = \Theta + W$

**LLMS Estimator:** $\hat{\Theta} = aX + b$

**Derivation:**
1.  **$Cov(X, \Theta)$:**
    $Cov(X, \Theta) = Cov(\Theta + W, \Theta)$
    $= Cov(\Theta, \Theta) + Cov(W, \Theta)$ (by linearity of covariance)
    Since $\Theta$ and $W$ are independent, $Cov(W, \Theta) = 0$.
    $Cov(X, \Theta) = Var(\Theta) = \sigma_{\Theta}^2$.

2.  **$Var(X)$:**
    $Var(X) = Var(\Theta + W)$
    Since $\Theta$ and $W$ are independent, $Var(X) = Var(\Theta) + Var(W) = \sigma_{\Theta}^2 + \sigma_W^2$.

3.  **$E[X]$:**
    $E[X] = E[\Theta + W] = E[\Theta] + E[W]$ (by linearity of expectation).

4.  **Coefficient 'a':**
    $a = \frac{Cov(X, \Theta)}{Var(X)} = \frac{\sigma_{\Theta}^2}{\sigma_{\Theta}^2 + \sigma_W^2}$.

5.  **Coefficient 'b':**
    $b = E[\Theta] - a \cdot E[X]$
    $b = E[\Theta] - \left(\frac{\sigma_{\Theta}^2}{\sigma_{\Theta}^2 + \sigma_W^2}\right) (E[\Theta] + E[W])$.

**General Form of Estimator:** $\hat{\Theta} = \left(\frac{\sigma_{\Theta}^2}{\sigma_{\Theta}^2 + \sigma_W^2}\right) X + \left(E[\Theta] - \frac{\sigma_{\Theta}^2}{\sigma_{\Theta}^2 + \sigma_W^2} (E[\Theta] + E[W])\right)$

```python
np.random.seed(44)
num_samples = 10000

# Define parameters for the multivariate normal distribution
# Mean vector [E[Theta], E[X1], E[X2]]
means = np.array([0, 0, 0])

# Covariance matrix for [Theta, X1, X2]
#       Theta X1 X2
# Theta [4,   1,  1]
# X1    [1,   4,  1]
# X2    [1,   1,  4]
cov_matrix = np.array([ [4, 1, 1], [1, 4, 1], [1, 1, 4]])

# Generate correlated random variables
simulated_data = np.random.multivariate_normal(means, cov_matrix, num_samples)

Theta_actual = simulated_data[:, 0]
X1_obs = simulated_data[:, 1]
X2_obs = simulated_data[:, 2]

# Apply the derived LLMS estimator
# hat_Theta = (1/5) * X1 + (1/5) * X2
a1_derived = 1/5
a2_derived = 1/5
b_derived = 0

Theta_estimated = a1_derived * X1_obs + a2_derived * X2_obs + b_derived

# Evaluate the estimator's performance using Mean Squared Error (MSE)
mse = np.mean((Theta_actual - Theta_estimated)**2)

# Compare with Var(Theta) for context
print(f"Variance of actual Theta: {np.var(Theta_actual):.4f}")
```

### Scenario 3:
LLMS estimator with two observations $X_1$ and $X_2$

Given the random variables $\Theta$, $X_1$, and $X_2$ with the following conditions
- $E[\Theta] = E[X_1] = E[X_2] = \mu_{common}$
- $Var(\Theta) = Var(X_1) = Var(X_2) = \sigma_{common}^2$
- $Cov(\Theta, X_1) = Cov(\Theta, X_2) = Cov(X_1, X_2) = \rho_{common}$

**LLMS Estimator:** $\hat{\Theta} = a_1 X_1 + a_2 X_2 + B$

**Derivation:**

1.  **Coefficient B:**
    $E[\Theta - \hat{\Theta}] = 0 \implies E[\Theta] - a_1 E[X_1] - a_2 E[X_2] - B = 0$
    $\mu_{common} - a_1 \mu_{common} - a_2 \mu_{common} - B = 0$
    $B = \mu_{common} (1 - a_1 - a_2)$

2.  **Coefficients $a_1$ and $a_2$ (Normal Equations):**
    $Cov(X_1, \Theta - \hat{\Theta}) = 0 \implies Cov(X_1, \Theta) - Cov(X_1, a_1 X_1 + a_2 X_2) = 0$
    $\rho_{common} - (a_1 Cov(X_1, X_1) + a_2 Cov(X_1, X_2)) = 0$
    $\rho_{common} - (a_1 \sigma_{common}^2 + a_2 \rho_{common}) = 0$
    $a_1 \sigma_{common}^2 + a_2 \rho_{common} = \rho_{common} \quad \text{(Equation 1)}

    $Cov(X_2, \Theta - \hat{\Theta}) = 0 \implies Cov(X_2, \Theta) - Cov(X_2, a_1 X_1 + a_2 X_2) = 0$
    $\rho_{common} - (a_1 Cov(X_2, X_1) + a_2 Cov(X_2, X_2)) = 0$
    $\rho_{common} - (a_1 \rho_{common} + a_2 \sigma_{common}^2) = 0$
    $a_1 \rho_{common} + a_2 \sigma_{common}^2 = \rho_{common} \quad \text{(Equation 2)}

    Solving Equation 1 and Equation 2:
    Subtracting Eq 2 from Eq 1:
    $(a_1 \sigma_{common}^2 + a_2 \rho_{common}) - (a_1 \rho_{common} + a_2 \sigma_{common}^2) = 0$
    $a_1 (\sigma_{common}^2 - \rho_{common}) - a_2 (\sigma_{common}^2 - \rho_{common}) = 0$
    $(a_1 - a_2)(\sigma_{common}^2 - \rho_{common}) = 0$

    This implies either $a_1 = a_2$ or $\sigma_{common}^2 = \rho_{common}$. Assuming $\sigma_{common}^2 \neq \rho_{common}$ (i.e., $X_1$ and $X_2$ are not perfectly correlated with $\Theta$ or each other), we have $a_1 = a_2$.

    Substitute $a_1 = a_2$ into Equation 1:
    $a_1 \sigma_{common}^2 + a_1 \rho_{common} = \rho_{common}$
    $a_1 (\sigma_{common}^2 + \rho_{common}) = \rho_{common}$
    $a_1 = \frac{\rho_{common}}{\sigma_{common}^2 + \rho_{common}}$

    Thus, $a_1 = a_2 = \frac{\rho_{common}}{\sigma_{common}^2 + \rho_{common}}$.

**General Form of Estimator:** $\hat{\Theta} = \left(\frac{\rho_{common}}{\sigma_{common}^2 + \rho_{common}}\right) X_1 + \left(\frac{\rho_{common}}{\sigma_{common}^2 + \rho_{common}}\right) X_2 + \mu_{common} \left(1 - \frac{2\rho_{common}}{\sigma_{common}^2 + \rho_{common}}\right)$

### Scenario 4:
LLMS estimator for a non-linear observation model $X = \Theta^n + W$

Given the random variables $\Theta$, $W$, and an observation $X$ with the following conditions:
- $\Theta$ and $W$ are independent.
- $E[W] = 0$ (zero-mean noise).
- $Var(W) = \sigma_W^2$.
- The observation is $X = \Theta^n + W$, where $n$ is an integer (e.g., $n=2$ for a quadratic relationship).
- $W$ is assumed to be relatively small compared to the range of $\Theta^n$.

**LLMS Estimator:** $\hat{\Theta} = aX + b$

**Derivation:**
The Linear Least Mean Squares (LLMS) estimator finds the best linear approximation of $\Theta$ given $X$. The coefficients $a$ and $b$ are always determined by the following general formulas, regardless of the underlying relationship being non-linear:

$a = \frac{Cov(X, \Theta)}{Var(X)}$
$b = E[\Theta] - a \cdot E[X]$

Let's derive the components needed for $a$ and $b$ based on our scenario:

1.  **$E[X]$:**
    $E[X] = E[\Theta^n + W]$
    By linearity of expectation, and since $E[W] = 0$:
    $E[X] = E[\Theta^n] + E[W] = E[\Theta^n]$.

2.  **$Cov(X, \Theta)$:**
    $Cov(X, \Theta) = E[X\Theta] - E[X]E[\Theta]$
    First, let's find $E[X\Theta]$:
    $E[X\Theta] = E[(\Theta^n + W)\Theta] = E[\Theta^{n+1} + W\Theta]$
    By linearity of expectation:
    $E[X\Theta] = E[\Theta^{n+1}] + E[W\Theta]$
    Since $\Theta$ and $W$ are independent and $E[W] = 0$, $E[W\Theta] = E[W]E[\Theta] = 0$.
    So, $E[X\Theta] = E[\Theta^{n+1}]$.
    Substituting back into the covariance formula:
    $Cov(X, \Theta) = E[\Theta^{n+1}] - E[\Theta^n]E[\Theta]$.

3.  **$Var(X)$:**
    $Var(X) = Var(\Theta^n + W)$
    Since $\Theta^n$ and $W$ are independent:
    $Var(X) = Var(\Theta^n) + Var(W) = Var(\Theta^n) + \sigma_W^2$.

**General Form of Estimator Coefficients:**

$a = \frac{E[\Theta^{n+1}] - E[\Theta^n]E[\Theta]}{Var(\Theta^n) + \sigma_W^2}$

$b = E[\Theta] - a \cdot E[\Theta^n]$

**Note on Higher-Order Moments:**
The crucial aspect of this scenario is that the calculation of $a$ and $b$ depends on higher-order moments of $\Theta$ (i.e., $E[\Theta]$, $E[\Theta^n]$, $E[\Theta^{n+1}]$, and $Var(\Theta^n)$). These moments are specific to the probability distribution of $\Theta$ and the value of $n$. For a general distribution of $\Theta$ and arbitrary $n$, these analytical calculations can become complex. In a practical setting, these values are often estimated empirically from data, or assumptions are made about $\Theta$'s distribution.

```python
import pandas as pd
import numpy as np

# Load the California Housing dataset
df_housing = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Select the variables for Theta, X1, and X2
# For this scenario, let's use 'median_house_value' as Theta,
# 'median_income' as X1, and 'housing_median_age' as X2.
Theta_actual_housing = df_housing['median_house_value']
X1_housing = df_housing['median_income']
X2_housing = df_housing['housing_median_age']

# Empirical statistics for the dataset
# Means
mean_Theta_h = Theta_actual_housing.mean()
mean_X1_h, mean_X2_h = X1_housing.mean(), X2_housing.mean()
# Variances
var_Theta_h = Theta_actual_housing.var()
var_X1_h, var_X2_h = X1_housing.var(), X2_housing.var()
# Covariances
cov_Theta_X1_h = Theta_actual_housing.cov(X1_housing)
cov_Theta_X2_h = Theta_actual_housing.cov(X2_housing)
cov_X1_X2_h = X1_housing.cov(X2_housing)

# Approximate 'common' parameters for Scenario 3's formula
# Note: In a real dataset, these values are rarely perfectly common.
# We average them to fit the structure of the Scenario 3 derivation.

mu_common_approx = np.mean([mean_Theta_h, mean_X1_h, mean_X2_h])
sigma_common_sq_approx = np.mean([var_Theta_h, var_X1_h, var_X2_h])
rho_common_approx = np.mean([cov_Theta_X1_h, cov_Theta_X2_h, cov_X1_X2_h])

# Use the previously derived formula for Scenario 3 to calculate coefficients
# The formula is: a1 = a2 = rho_common / (sigma_common_sq + rho_common)
# B = mu_common * (1 - a1 - a2)

# Ensure the denominator is not zero
if (sigma_common_sq_approx + rho_common_approx) != 0:
    a1_housing = rho_common_approx /(sigma_common_sq_approx + rho_common_approx)
    a2_housing = a1_housing
else:
    a1_housing, a2_housing = 0.0, 0.0

B_housing = mu_common_approx * (1 - a1_housing - a2_housing)

# Apply the estimator to the dataset
Theta_estimated_housing = a1_housing * X1_housing + a2_housing * X2_housing + B_housing

# Evaluate the estimator's performance using Mean Squared Error (MSE)
mse_housing = np.mean((Theta_actual_housing - Theta_estimated_housing)**2)
var_Theta_actual_housing = np.var(Theta_actual_housing)
```

```python
np.random.seed(42)
num_samples = 10000

# Define parameters for the simulation
n = 2  # Non-linear power for Theta

# Parameters for Theta (e.g., Uniform distribution)
theta_min, theta_max = 1, 10 # Theta values between 1-10 to avoid issues with Theta^2 around 0
Theta_actual_s4 = np.random.uniform(theta_min, theta_max, num_samples)

# Parameters for noise W
sigma_W_s4 = 2.0  # Standard deviation of noise
W_s4 = np.random.normal(0, sigma_W_s4, num_samples)

# Generate X based on the non-linear relationship
X_s4 = Theta_actual_s4**n + W_s4

# Calculate empirical statistics from simulated data
E_Theta_s4 = np.mean(Theta_actual_s4)
E_X_s4 = np.mean(X_s4)
Cov_X_Theta_s4 = np.cov(X_s4, Theta_actual_s4)[0, 1] # Extract covariance
Var_X_s4 = np.var(X_s4)
Var_Theta_s4 = np.var(Theta_actual_s4)

# Calculate 'a' and 'b' for the LLMS estimator
# Ensure Var_X is not zero to prevent division by zero
if Var_X_s4 > 1e-9:
    a_s4 = Cov_X_Theta_s4 / Var_X_s4
else:
    a_s4 = 0.0 # If X has no variance, it's a constant

b_s4 = E_Theta_s4 - a_s4 * E_X_s4

# Apply the estimator
Theta_estimated_s4 = a_s4 * X_s4 + b_s4

# Evaluate the estimator's performance
mse_s4 = np.mean((Theta_actual_s4 - Theta_estimated_s4)**2)
print(f"\nMean Squared Error (MSE) of the LLMS estimator: {mse_s4:.3f}")
print(f"Variance of actual Theta: {Var_Theta_s4:.3f}")
```
