### Probabilistic Methods - Bayesian Inference and Methods of Estimations

### **Operations on Random Variables**

Sums of Independent R.V.s - Mathematical derivations for the Probability Mass Functions (PMF) and Probability Density Functions (PDF) of combined discrete and continuous variables. Conditioning - Techniques for conditioning multiple continuous random variables on specific events.

####**Discrete Random Variables**

Let $X, Y$ be discrete random variables and independent

$Z = X + Y = g(X, T)$

$P_Z(z) = \sum_x P(X=x, Y=z-x) = \sum_x P_X(x) P_Y(z-x)$

####**Continuous Random Variables**

Let $X, Y$ be continuous random variables and independent

$Z = X + Y$

$f_Z(z) = \int_{-\infty}^{\infty} f_X(x) f_Y(z-x) \, dx$$

####**Conditioning on an event - Multiple continuous r.v.s**

For discrete and continuous r.v.s
- $p_X(x) = P(X = x), f_X(x) \Delta \approx P(x \le X \le x + \Delta)$.

**Conditional on event**
- $A$ $p_{X|A}(x) = P(X = x \mid A), f_{X|A}(x) \Delta \approx P(x \le X \le x + \Delta \mid A)$.

**Probability in Region**
- $B$, $P(X \in B) = \sum_{x \in B} p_X(x), P(X \in B) = \int_B f_X(x) \, dx$.

**Conditional Probability in Region**
- $B$, $P(X \in B \mid A) = \sum_{x \in B} p_{X|A}(x), P(X \in B \mid A) = \int_B f_{X|A}(x) \, dx$

Normalization Property
- $\sum_x p_{X|A}(x) = 1, \int_{-\infty}^{\infty} f_{X|A}(x) \, dx = 1$.

**Conditional PDF of $X$, given that $X \in A$**

The probability density of $X$ restricted to a specific set $A$ is $P(x \le X \le x + \Delta \mid X \in A) \approx f_{X \mid X \in A}(x) \Delta$.

By the definition of conditional probability $\frac{P(x \le X \le x + \Delta, X \in A)}{P(A)} = \frac{P(x \le X \le x + \Delta)}{P(A)} \approx \frac{f_X(x) \Delta}{P(A)}$

The resulting conditional PDF is defined as $f_{X \mid X \in A}(x) =
\begin{cases}
\frac{f_X(x)}{P(A)} & \text{if } x \in A \\
0 & \text{if } x \notin A
\end{cases}$

**Conditional Expectation of $X$ Given an Event**

Unconditional Expectation $E[X] = \sum_{x} x p_X(x), \quad E[X] = \int_{-\infty}^{\infty} x f_X(x) \, dx$

Conditional Expectation given event $A$ $E[X \mid A] = \sum_{x} x p_{X|A}(x), \quad E[X \mid A] = \int_{-\infty}^{\infty} x f_{X|A}(x) \, dx$

**Expected Value Rule (Law of the Unconscious Statistician)**

Unconditional $E[g(X)] = \sum_{x} g(x) p_X(x), \quad E[g(X)] = \int_{-\infty}^{\infty} g(x) f_X(x) \, dx$


Conditional given event $A$, $E[g(X) \mid A] = \sum_{x} g(x) p_{X|A}(x), \quad E[g(X) \mid A] = \int_{-\infty}^{\infty} g(x) f_{X|A}(x) \, dx$

####**Memorylessness of the Exponential PDF**

Probabilistic model: Exponential light bulb

Bulb lifetime $T$: $\text{Exponential}(\lambda)$
- $P(T > x) = e^{-\lambda x} \quad \text{for } x \ge 0$
- $T > t \quad$ Random variable $X = T - t$ (remaining lifetime)
- $P(X > x \mid T > t) = e^{-\lambda x}, \quad \text{for } x \ge 0$

Derivation:


$\begin{aligned}
P(X > x \mid T > t) &= \frac{P(T - t > x, T > t)}{P(T > t)} = \frac{P(T > t + x, T > t)}{P(T > t)} = \frac{P(T > t + x)}{P(T > t)}
\end{aligned}$

$\begin{aligned}
P(X > x \mid T > t) &= \frac{e^{-\lambda(t + x)}}{e^{-\lambda t}} = e^{-\lambda x}
\end{aligned}$

### **Expectations & Theorems**

Total Probability & Expectations - Implementation of the Law of Total Probability and the Total Expectation Theorem across partitions. Expected Value Rule - Application of the "Law of the Unconscious Statistician" for both unconditional and conditional cases.

####**Total Probability and Expectations Theorem**

The law of total probability and the law of total expectation across partitions $A_1, \dots, A_n$.
- $P(B) = P(A_1)P(B \mid A_1) + \dots + P(A_n)P(B \mid A_n)$
- $p_X(x) = P(A_1)p_{X|A_1}(x) + \dots + P(A_n)p_{X|A_n}(x)$
- $f_X(x) = P(A_1)f_{X|A_1}(x) + \dots + P(A_n)f_{X|A_n}(x)$
- $E[X] = P(A_1)E[X \mid A_1] + \dots + P(A_n)E[X \mid A_n]$

####**Joint Continuous Random Variables and Joint PDFs**

Probability Density and Mass Functions
- Marginal: $p_X(x)$, $f_X(x)$ & Joint: $p_{X,Y}(x,y)$, $f_{X,Y}(x,y)$

Non-negativity
- $p_{X,Y}(x,y) = P(X=x, Y=y) \ge 0, f_{X,Y}(x,y) \ge 0$

For discrete variables and continuous variables
- $P((X,Y) \in B) = \sum \sum_{(x,y) \in B} p_{X,Y}(x,y)$ & $P((X,Y) \in B) = \iint_B f_{X,Y}(x,y) \, dx \, dy$

Normalization Property
- $\sum_x \sum_y p_{X,Y}(x,y) = 1$ & $\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{X,Y}(x,y) \, dx \, dy = 1$

### Mixed Distributions

####**Independence**

Discrete and Continuous Definitions
- $p_{X,Y}(x,y) = p_X(x) p_Y(y) \text{ for all } x, y$
- $f_{X,Y}(x,y) = f_X(x) f_Y(y) \text{ for all } x, y$

Conditional Relationship (joint density) $f_{X,Y}(x,y) = f_{X|Y}(x|y) f_Y(y)$


Independence $f_{X|Y}(x|y) = f_X(x) \text{ for all } y \text{ with } f_Y(y) > 0 \text{ and all } x$

Properties of Independent Variables - If $X$ and $Y$ are independent
- $E[XY] = E[X] E[Y]$, $\text{var}(X + Y) = \text{var}(X) + \text{var}(Y)$

If $X$ and $Y$ are independent, $g(X)$ and $h(Y)$ are also independent
- $E[g(X)h(Y)] = E[g(X)] E[h(Y)]$

####**Independent Normal Random Variables**

Let $X, Y$ be independent normal random variables $X \sim N(\mu_x, \sigma_x^2), Y \sim N(\mu_y, \sigma_y^2)$, $Z = X + Y$

The sum of finitely many independent normals is normal
- $f_Z(z) = \int_{-\infty}^{\infty} f_X(x) f_Y(z-x) \, dx$

Probability Density Functions $f_X(x) = \frac{1}{\sqrt{2\pi}\sigma_x} e^{-\frac{(x-\mu_x)^2}{2\sigma_x^2}}$ $f_Y(y) = \frac{1}{\sqrt{2\pi}\sigma_y} e^{-\frac{(y-\mu_y)^2}{2\sigma_y^2}}$

**Convolution Result**

- $f_Z(z) = \int_{-\infty}^{\infty} f_X(x) f_Y(z-x) \, dx$, $f_Z(z) = \frac{1}{\sqrt{2\pi(\sigma_x^2 + \sigma_y^2)}} \exp\left( -\frac{(z - \mu_x - \mu_y)^2}{2(\sigma_x^2 + \sigma_y^2)} \right) \sim N(\mu_x + \mu_y, \sigma_x^2 + \sigma_y^2)$


####**Covariance and Variance Properties**


- $\text{cov}(X, Y) = E[(X - E[X])(Y - E[Y])], X, Y$ independent, $\text{cov}(X, Y) = 0$.


Covariance Properties

- Relationship to Variance: $\text{cov}(X, X) = E[X^2] - (E[X])^2 = \text{var}(X)$
- Linearity: $\text{cov}(aX + b, Y) = a\,\text{cov}(X, Y)$
- Additivity: $\text{cov}(X, Y + Z) = \text{cov}(X, Y) + \text{cov}(X, Z)$

####**The Variance of a Sum of Random Variables**

For two variables, $\begin{aligned} \text{var}(X_1 + X_2) &= \text{var}(X_1) + \text{var}(X_2) + 2\,\text{cov}(X_1, X_2) \end{aligned}$

For $n$ variables $\text{var}\left(\sum_{i=1}^{n} X_i\right) = \sum_{i=1}^{n} \text{Var}(X_i) + \sum_{i \neq j} \text{cov}(X_i, X_j)$

####**The Correlation Coefficient**

Dimensionless version of covariance $\rho(X,Y) = \frac{\text{cov}(X,Y)}{\sigma_X \sigma_Y}$

Measure of the degree of association between $X$ and $Y$

- Independent $\rightarrow \rho=0$, "uncorrelated"
- $\rho(X,X) = \frac{\text{var}(X)}{\sigma_X^2} = 1$

**Linear Relationships**

If $|\rho| = 1 \iff (X - E[X]) = C(Y - E[Y])$ (linearly related).

**Scaling Properties**

Given the covariance property $\text{cov}(aX + b, Y) = a \cdot \text{cov}(X, Y)$, the correlation transforms as $\rho(aX + b, Y) = \frac{a \cdot \text{cov}(X, Y)}{|a| \sigma_X \sigma_Y}$.

####**The Conditional Variance as a Random Variable**

The variance of $X$ is defined as
- $\text{var}(X) = E[(X - E[X])^2]$

Conditional Variance given $Y = y$
- $\text{var}(X \mid Y = y) = E[(X - E[X \mid Y = y])^2 \mid Y = y]$

$\text{var}(X \mid Y)$ is the random variable that takes the value
- $\text{var}(X \mid Y = y)$ when $Y = y$.

Law of Total Variance
- $\text{var}(X) = E[\text{var}(X \mid Y)] + \text{var}(E[X \mid Y])$.



####**Sum of a random number of independent r.v.s**

Let $N$ be a random variable representing, for example, the number of stores visited, and $X_i$ be the money spent in store $i$. Assume $X_i$ are independent and identically distributed (i.i.d.) and are independent of $N$.Let $Y$ be the total sum:$Y = X_1 + \dots + X_N$

Conditional Expectation $E[Y \mid N = n] = E[X_1 + \dots + X_n \mid N = n] = E[X_1 + \dots + X_n] = n E[X]$

Total Expectation Theorem $E[Y] = \sum_{n} p_N(n) E[Y \mid N = n] = E[N] E[X]$

Law of iterated Expectations $E[Y] = E[E[Y \mid N]] = E[N E[X]] = E[N] E[X]$

Variance of Sum of Independent r.v's. Given $Y = X_1 + \dots + X_N$, we use the Law of Total Variance
- $\operatorname{var}(Y) = E[\operatorname{var}(Y \mid N)] + \operatorname{var}(E[Y \mid N])$
- $E[\operatorname{var}(Y \mid N)] = E[N \operatorname{var}(X)] = E[N] \operatorname{var}(X)$

- $\operatorname{var}(Y) = E[N] \operatorname{var}(X) + (E[X])^2 \operatorname{var}(N)$

### Bayesian Inference

####**The Bayes Rule: A Theme with Variations**

For a random variable $X$:
- Unobserved value $x$ (prior $p_X(\cdot)$)
- Observed value $y$ (model $p_{Y|X}(\cdot \mid \cdot)$)
- Posterior $p_{X|Y}(\cdot \mid y)$

**Joint Probability and Bayes Theorem**

The relationship between the joint and conditional probabilities is given by
- $p_{X,Y}(x,y) = p_X(x) p_{Y|X}(y|x) = p_Y(y) p_{X|Y}(x|y)$

Rearranging to find the posterior
- $p_{X|Y}(x|y) = \frac{p_X(x) p_{Y|X}(y|x)}{p_Y(y)}$

Where the marginal $p_Y(y)$ is calculated as
- $p_Y(y) = \sum_{x'} p_X(x') p_{Y|X}(y|x')$

**Continuous Case (Probability Density Functions)**

Similarly, for continuous variables $f_{X,Y}(x,y) = f_X(x) f_{Y|X}(y|x)$

The posterior density is $f_{X|Y}(x|y) = \frac{f_X(x) f_{Y|X}(y|x)}{f_Y(y)}$

Where the marginal $f_Y(y)$ is calculated via integration $f_Y(y) = \int f_X(x') f_{Y|X}(y|x') \, dx'$



####**The Bayes Rule: One Discrete and One Continuous Random Variable**

$K$  discrete random variable, $Y$  continuous random variable.

For $\Delta > 0$ and $\Delta \approx 0$

$P(K = k, y \le Y \le y + \Delta) = P(K = k) P(y \le Y \le y + \Delta \mid K = k) \approx p_K(k) f_{Y|K}(y \mid k) \Delta =$

$P(y \le Y \le y + \Delta) P(K = k \mid y \le Y \le y + \Delta) \approx f_Y(y) \Delta p_{K|Y}(k \mid y)$

Finally:

$p_{K|Y}(k \mid y) = \frac{p_K(k) f_{Y|K}(y \mid k)}{f_Y(y)}, f_{Y|K}(y \mid k) = \frac{f_Y(y) p_{K|Y}(k \mid y)}{p_K(k)}$

Marginal Distributions

$f_Y(y) = \sum_{k'} p_K(k') f_{Y|K}(y \mid k')$, $p_K(k) = \int f_Y(y') p_{K|Y}(k \mid y') \, dy'$



####**Bayes' Rule: Discrete Unknown, Continuous Measurement**

Unknown $K$: Equally likely to be $-1$ or $+1$.

Measurement $Y$: $Y = K + W$, where $W \sim {N}(0,1)$.

Likelihood Distributions
- $Y \mid K=1 \sim {N}(1,1)$, $Y \mid K=-1 \sim {N}(-1,1)$

Posterior Probability
- For $K=1$ given $Y=y$, denoted as $P_{K \mid Y}(1 \mid y)$, we use $p_{K \mid Y}(k \mid y) = \frac{p_K(k) f_{Y \mid K}(y \mid k)}{f_Y(y)}$
- Where the marginal density $f_Y(y)$ is $f_Y(y) = \sum_{k'} p_K(k') f_{Y \mid K}(y \mid k')$

Components & Derivation
- Prior: $p_K(k) = \frac{1}{2}$ for $k \in \{-1, 1\}$
- Likelihood: $f_{Y \mid K}(y \mid k) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{1}{2}(y - k)^2\right)$

Marginal Density
- $f_Y(y) = \frac{1}{2} \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{1}{2}(y+1)^2\right) + \frac{1}{2} \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{1}{2}(y-1)^2\right)$

Posterior Result
- $p_{K \mid Y}(1 \mid y) = \frac{1}{1 + \exp(-2y)}$

Two conditional probability density functions $f_{Y \mid K}(y \mid 1) \text{ and } f_{Y \mid K}(y \mid -1)$, and the posterior probability $P(K=1 \mid Y=y)$.

### Inference

####**The Bayesian Inference Framework**

- Unknown parameter $\Theta$: Treated as a r.v. with a prior distribution $p_{\Theta}$ (PMF) or $f_{\Theta}$ (PDF).

- Observation $X$: Defined by the observation model $p_{X \mid \Theta}$ or $f_{X \mid \Theta}$.

- Bayes' Rule: Used to find $p_{\Theta \mid X}(\theta \mid X=x)$ or $f_{\Theta \mid X}(\theta \mid X=x)$.

- Posterior Distribution: The resulting PMF $p_{\Theta \mid X}(\theta \mid x)$ or PDF $f_{\Theta \mid X}(\theta \mid x)$.

- Point Estimation:
    - Estimate: $\hat{\theta} = g(x)$ (a specific numerical value).
    - Estimator: $\hat{\Theta} = g(X)$ (a random variable).

- Maximum A Posteriori (MAP):
    - $p_{\Theta \mid X}(\theta^* \mid x) = \max_{\theta} p_{\Theta \mid X}(\theta \mid x), f_{\Theta \mid X}(\theta^* \mid x) = \max_{\theta} f_{\Theta \mid X}(\theta \mid x)$

- Conditional Expectation: $E[\Theta \mid X = x]$ (also known as LMS  Least Mean Squares).

####**Discrete $\Theta$, Discrete $X$**

Posterior Probability (Bayes' Rule): $P_{\Theta|X}(\theta \mid x) = \frac{p_{\Theta}(\theta) \cdot p_{X|\Theta}(x \mid \theta)}{p_{X}(x)}$

Marginal Probability (Law of Total Probability): $p_{X}(x) = \sum_{\theta'} p_{\Theta}(\theta') p_{X|\Theta}(x \mid \theta')$

Conditional Probability of Error: $P(\hat{\Theta} \neq \Theta \mid X = x)$

Overall Probability of Error: $P(\hat{\Theta} \neq \Theta) = \sum_{x} P(\hat{\Theta} \neq \Theta \mid X = x) p_{X}(x)$  $= \sum_{\theta} P(\hat{\Theta} \neq \theta \mid \Theta = \theta) p_{\Theta}(\theta)$

Key Rules: **MAP** rule (Maximum A Posteriori) & **LMS** (Least Mean Squares)

####**Discrete $\Theta$, Continuous $X$**

Posterior Distribution:$p_{\Theta|X}(\theta \mid x) = \frac{p_{\Theta}(\theta) f_{X|\Theta}(x \mid \theta)}{f_X(x)}$

Marginal Density: $f_X(x) = \sum_{\theta'} p_{\Theta}(\theta') f_{X|\Theta}(x \mid \theta')$

Conditional Probability of Error: $P(\hat{\theta} \neq \Theta \mid X = x)$

Overall Probability of Error: $P(\hat{\Theta} \neq \Theta) = \int P(\hat{\Theta} \neq \Theta \mid X = x) f_X(x) \, dx$  &

$P(\hat{\Theta} \neq \Theta) = \sum_{\theta} P(\hat{\Theta} \neq \theta \mid \Theta = \theta) p_{\Theta}(\theta)$

####**Continuous $\Theta$, Continuous $X$**

The posterior density is given by Bayes' Rule $f_{\Theta|X}(\theta \mid x) = \frac{f_{\Theta}(\theta) f_{X|\Theta}(x \mid \theta)}{f_X(x)}$.

Where the marginal density of $X$ is $f_X(x) = \int_{-\infty}^{\infty} f_{\Theta}(\theta') f_{X|\Theta}(x \mid \theta') \, d\theta'$.

Linear Normal Models: Estimation of a Noisy Signal

Assuming $X = \Theta + W$, where $\Theta$ and $W$ are independent normal random variables.
- Multi-dimensional versions: Extension to many normal parameters and multiple observations.

Estimating the Parameter of a Uniform
- $X \sim \text{Unif}[0, \Theta], \quad \Theta \sim \text{Unif}[0, 1]$

Estimation Performance
- Estimator: $\hat{\Theta} = g(X)$ (e.g., MAP or LMS)
- Metrics: Conditional Bias: $E[(\hat{\Theta} - \Theta) \mid X = x]$, and MSE: $E[(\hat{\Theta} - \Theta)^2]$

####**Summary of Bayesian Inference**

- Problem data: $p_{\Theta}(\cdot)$, $p_{Y|\Theta}(\cdot|\cdot)$
- Given the value $x$ of $X$: Find $p_{\Theta|X}(\cdot|x)$ using the appropriate version of the Bayes rule.
- Estimator $\hat{\Theta} = g(X)$, Estimate $\hat{\theta} = g(x)$
    - MAP: $\hat{\theta}_{MAP} = g_{MAP}(x)$ maximizes $p_{\Theta|X}(\theta |x)$
    - LMS: $\hat{\theta}_{LMS} = g_{LMS}(x) = E[\Theta \mid X = x]$
- Performance evaluation of an estimator $\hat{\Theta}$
    - $P(\hat{\Theta} \neq \Theta \mid X = x)$; $E[(\hat{\Theta} - \Theta)^2 \mid X = x]$
    - $P(\hat{\Theta} \neq \Theta)$; $E[(\hat{\Theta} - \Theta)^2]$  Total probability expectation.

####**Inferring the Unknown Bias of a Coin and the Beta Distribution**

Standard Example
- Coin with bias $\theta$; prior $f_{\Theta}(\theta)$; Fix $n$; $K = \text{number of heads}$. Assume $f_{\Theta}(\theta)$ is uniform in $[0, 1]$

This is a Beta distribution with parameters $(k + 1, n - k + 1)$.
- $f_{\Theta|K}(\theta | k) = 1 \cdot \binom{n}{k} \theta^k (1 - \theta)^{n-k} = \frac{1}{d(n, k)} \theta^k (1 - \theta)^{n-k} \quad \text{for } \theta \in [0, 1]$

General Beta Prior

- If the prior is Beta: $f_{\Theta}(\theta) = \frac{1}{c} \theta^{\alpha} (1 - \theta)^{\beta}, \quad \alpha, \beta \geq 0$

- Posterior distribution: $f_{\Theta|K}(\theta | k) = \frac{\frac{1}{c} \theta^{\alpha} (1 - \theta)^{\beta} \binom{n}{k} \theta^k (1 - \theta)^{n-k}}{P_K(k)} = d \theta^{\alpha + k} (1 - \theta)^{\beta + n - k}$

Assume $f_{\Theta}(\theta)$ is uniform on $[0, 1]$ (Uniform Prior Assumption).

Posterior Distribution: $f_{\Theta|K}(\theta | k) = \frac{1}{d(n,k)} \theta^k (1 - \theta)^{n - k}$

Posterior Expectation (LMS Estimate):
- $E[\Theta \mid K = k] = \int_{0}^{1} \theta f_{\Theta|K}(\theta | k) d\theta = \frac{k + 1}{n + 2}$ (Note: For large $n$, this approximates to $\frac{k}{n}$ )

**MAP Estimate** is $\hat{\theta}_{MAP} = \frac{k}{n}$

Estimating a Normal Random Variable with Additive Normal Noise

Let $\Theta, W \sim N(0,1)$ be independent.$X = \Theta + W$. Given $\Theta = \theta$, then $X = \theta + W \sim N(\theta, 1)$
- $f_{X|\Theta}(x \mid \theta) = c(x) e^{-\text{quadratic}(\theta)}$

**Estimators**:
- $\hat{\theta}_{MAP} = \hat{\theta}_{LMS} = E[\Theta \mid X = x] = \frac{x}{2}$
- $\hat{\Theta}_{MAP} = \hat{\Theta}_{LMS} = E[\Theta \mid X] = \frac{X}{2}$

Even with general means and variances, the posterior distribution remains Normal.

####**The Case of Multiple Observations**

Assume $X_1, \dots, X_n$ are observations such that $X_i = \Theta + W_i, \Theta \sim N(x_0, \sigma_0^2), W_i \sim N(0, \sigma_i^2)$

where $\Theta, W_1, \dots, W_n$ are independent.


**Conditional Densities** - Given $\Theta = \theta$, then $X_i = \theta + W_i \sim N(\theta, \sigma_i^2)$.

The likelihood for a single observation is $f_{X_i|\Theta}(x_i \mid \theta) = c_i \exp\left( -\frac{(x_i - \theta)^2}{2\sigma_i^2} \right)$

For multiple observations, since $W_i$ are independent, the $X_i$ are conditionally independent given $\Theta:$
- $f_{\mathbf{X}|\Theta}(\mathbf{x} \mid \theta) = f_{X_1, \dots, X_n|\Theta}(x_1, \dots, x_n \mid \theta) = \prod_{i=1}^{n} f_{X_i|\Theta}(x_i \mid \theta)$

**Posterior Distribution** - Using Bayes' Rule, the posterior density is:
- $f_{\Theta|\mathbf{X}}(\theta \mid \mathbf{x}) = \frac{1}{f_{\mathbf{X}}(\mathbf{x})} \cdot C_0 \exp\left( -\frac{(\theta - x_0)^2}{2\sigma_0^2} \right) \cdot \prod_{i=1}^{n} c_i \exp\left( -\frac{(x_i - \theta)^2}{2\sigma_i^2} \right)$

This distribution is proportional to the exponential of a quadratic in $\theta$, signifying a Normal distribution  
- $f_{\Theta|\mathbf{X}}(\theta \mid \mathbf{x}) = C \exp(-\text{quad}(\theta))$

**Optimization and Estimation** - Set the derivative of the quadratic exponent to zero
- $\frac{d}{d\theta} \text{quad}(\theta) = 0 \implies \theta \sum_{i=0}^{n} \frac{1}{\sigma_i^2} = \sum_{i=0}^{n} \frac{x_i}{\sigma_i^2}$

The MAP and LMS estimates coincide because the posterior is symmetric and unimodal (Gaussian)
- $\hat{\theta}_{\text{MAP}} = \hat{\theta}_{\text{LMS}} = E[\Theta \mid \mathbf{X}=\mathbf{x}] = \frac{\sum_{i=0}^{n} \frac{x_i}{\sigma_i^2}}{\sum_{i=0}^{n} \frac{1}{\sigma_i^2}}$



####**Mean Square Error**

The posterior distribution is given by
- $f_{\Theta|X}(\theta|x) = c \cdot \exp\{-Q(\theta)\}$

Where $X_i = \Theta + W_i$, and the estimate is
- $\hat{\theta} = \frac{\sum_{i=0}^{n} \frac{x_i}{\sigma_i^2}}{\sum_{i=0}^{n} \frac{1}{\sigma_i^2}}$

The quadratic term $Q(\theta)$ is defined as  $Q(\theta) = \sum_{i=0}^{n} \frac{(\theta - x_i)^2}{2\sigma_i^2}$

**Performance Measures**

Conditional Mean Square Error
- $E[(\Theta - \hat{\Theta})^2 \mid X=x] = E[(\Theta - \hat{\theta})^2 \mid X=x] = \text{Var}(\Theta \mid X=x) = \frac{1}{\sum_{i=0}^{n} \frac{1}{\sigma_i^2}}$

Unconditional Mean Square Error
- $E[(\Theta - \hat{\Theta})^2] = \int E[(\Theta - \hat{\Theta})^2 \mid X=x] f_X(x) \, dx = \frac{1}{\sum_{i=0}^{n} \frac{1}{\sigma_i^2}}$


The marginal density $f_X(x) = c \cdot e^{-(\alpha x^2 + \beta x + \gamma)}$


Where $\alpha > 0$. This normal distribution with $\mu = -\frac{\beta}{2\alpha}, \sigma^2 = \frac{1}{2\alpha}$

The precision term $\alpha$ defined as $\alpha = \sum_{i=0}^{n} \frac{1}{2\sigma_i^2}$


**Observations**

- Sensitivity: If some $\sigma_i^2$ are small, the MSE is small. If all $\sigma_i^2$ are large, the MSE is large.
- Equal Variance Case: When $\sigma_0^2 = \sigma_1^2 = \dots = \sigma_n^2 = \sigma^2$
    - MSE $\frac{1}{(n+1)/\sigma^2} = \frac{\sigma^2}{n+1}$
- Uniformity: The conditional mean squared error is the same for all $x$.
- Example Case: For $X = \Theta + W$, where $\Theta \sim N(0,1)$ and $W \sim N(0,1)$ are independent
    - $\hat{\Theta} = \frac{X}{2}, \quad E[(\Theta - \hat{\Theta})^2 \mid X=x] = \frac{1}{2}$


####**A Model with Normality Assumptions**

$X_i = \theta_0 + \theta_1 t_i + \theta_2 t_i^2 + W_i, \quad i = 1, \dots, n$

In general case:
- $f_{\Theta|X}(\theta \mid x) = \frac{f_{\Theta}(\theta) f_{X|\Theta}(x \mid \theta)}{f_X(x)}$
- $f_X(x) = \int f_{\Theta}(\theta) f_{X|\Theta}(x \mid \theta) \, d\theta$$


**Distributional Assumptions**

Assume $\Theta_j \sim N(0, \sigma_j^2)$ and $W_i \sim N(0, \sigma^2)$, all independent.

Given $\Theta = \theta = (\theta_0, \theta_1, \theta_2)$, then $X_i \sim N(\theta_0 + \theta_1 t_i + \theta_2 t_i^2, \sigma^2)$:
- $f_{X_i|\Theta}(x_i \mid \theta) = c \exp\left\{ -\frac{(x_i - \theta_0 - \theta_1 t_i - \theta_2 t_i^2)^2}{2\sigma^2} \right\}$

**Posterior Distribution**
- $f_{\Theta|X}(\theta \mid x) = \frac{1}{f_X(x)} \prod_{j=0}^{2} f_{\Theta_j}(\theta_j) \prod_{i=1}^{n} f_{X_i|\Theta}(x_i \mid \theta)$
- $f_{\Theta|X}(\theta \mid x) \propto \exp\left\{ -\frac{1}{2} \left( \frac{\theta_0^2}{\sigma_0^2} + \frac{\theta_1^2}{\sigma_1^2} + \frac{\theta_2^2}{\sigma_2^2} \right) - \frac{1}{2\sigma^2} \sum_{i=1}^{n} (x_i - \theta_0 - \theta_1 t_i - \theta_2 t_i^2)^2 \right\}$


**MAP Estimate**

Maximize over $(\theta_0, \theta_1, \theta_2)$ by minimizing the quadratic function:
- $\frac{\partial Q(\theta)}{\partial \theta_j} = 0 \quad \text{(3 equations, 3 unknowns, linear system)}$



####**Summary Linear normal models**

- $\Theta_j$ and $X_i$ are linear functions of independent normal random variables.
- $f_{\Theta|X}(\theta \mid x) = c(x) \exp\{-\text{quadratic}(\theta_1, \dots, \theta_n)\}$  Linear Regression

**MAP estimate** Maximize over $(\theta_1, \dots, \theta_n)$

- Marginal posterior PDF of $\Theta_j$: $f_{\Theta_j|X}(\theta_j \mid x)$ is Normal.

####**Least Mean Squares (LMS) Estimation**

Minimize conditional mean squared error $E[(\Theta - \hat{\theta})^2 \mid X = x]$

General estimation model $\hat{\theta} = E[\Theta \mid X = x]$

**LMS Estimation in the Absence of Observations**

- Unknown $\Theta$, prior $p_{\Theta}(\theta)$:Interested in a point estimate $\hat{\theta}$ when no observations are available. MAP rule

- Conditional expectations

Criterion: Mean Squared Error (MSE) $E[(\Theta - \hat{\theta})^2]$

**Minimize Mean Squared Error**

To minimize $E[(\Theta - \hat{\theta})^2]$, the optimal estimate is $\hat{\theta} = E[\Theta]$

Optimal Mean Squared Error $E[(\Theta - E[\Theta])^2] = \text{var}(\Theta)$

####**LMS Estimation of $\Theta$ based on $X$**


- Unknown Parameter: $\Theta$ with prior PDF $p_{\Theta}(\theta)$.
- Goal: Find a point estimate $\hat{\theta}$.
- Observation: $X$ with model $p_{X|\Theta}(x|\theta)$. We observe $X = x$.

**Minimizing Mean Squared Error (MSE)**

Unconditional MSE: To minimize $E[(\Theta - \hat{\theta})^2]$, the optimal constant estimate is
- $\hat{\theta} = E[\Theta]$

Conditional MSE: To minimize $E[(\Theta - \hat{\theta})^2 \mid X = x]$, the optimal estimate is
- $\hat{\theta} = E[\Theta \mid X = x]$

**The LMS Estimator**

- LMS Estimate: $\hat{\theta} = E[\Theta \mid X = x]$
- LMS Estimator: $\hat{\Theta} = E[\Theta \mid X]$

**Key Properties**
- $E[\Theta]$ minimizes $E[(\Theta - \hat{\theta})^2]$.
- $E[\Theta \mid X = x]$ minimizes $E[(\Theta - \hat{\theta})^2 \mid X = x]$.
- The estimator $\hat{\Theta}_{\text{LMS}} = E[\Theta \mid X]$ minimizes the overall MSE, $E[(\Theta - g(X))^2]$, over all possible estimators $\hat{\Theta} = g(X)$.




####**LMS Estimation with Multiple Observations or Unknowns**:

- Unknown: $\Theta$; prior $p_{\Theta}(\theta)$; interested in a point estimate $\hat{\theta}$
- Observations: $X = (X_1, X_2, \dots, X_n)$; model $p_{X|\Theta}(x | \theta)$. Observe that $X = x$ ; Condition on $X = x$


LMS estimate
- $E[\Theta \mid X_1 = x_1, \dots, X_n = x_n]$

If $\Theta$ is a vector, apply to each component separately
- $\Theta = (\Theta_1, \dots, \Theta_m)$
- $\hat{\Theta}_j = E[\Theta_j \mid X_1 = x_1, \dots, X_n = x_n]$

Properties of the Estimation Error in LMS Estimation
- Estimator: $\hat{\Theta} = E[\Theta \mid X]$
- Error: $\tilde{\Theta} = \hat{\Theta} - \Theta$


Key Properties:
1. Mean of the error: $E[\tilde{\Theta} \mid X = x] = 0$
2. Orthogonality/Uncorrelatedness: $\text{cov}(\hat{\Theta}, \tilde{\Theta}) = 0$
3. Variance Decomposition: $\text{var}(\Theta) = \text{var}(\hat{\Theta}) + \text{var}(\tilde{\Theta})$


####**LMS Performance Evaluation**

LMS Estimate and Estimator

- LMS estimate: $\hat{\theta} = E[\Theta \mid X = x]$
- Estimator: $\hat{\Theta} = E[\Theta \mid X]$


**Expected Performance (Post-Measurement)**

Once we have a measurement, the performance is measured by the Mean Squared Error (MSE) conditioned on the specific observation:
- $MSE = E\left[ (\Theta - E[\Theta \mid X = x])^2 \mid X = x \right] = \text{var}(\Theta \mid X = x)$

The overall expected performance (the "design" phase) is the average MSE over all possible measurements
- $MSE = E\left[ (\Theta - E[\Theta \mid X])^2 \right] = E[\text{var}(\Theta \mid X)]$


**Challenges in LMS Estimation**

Full correct model $f_{X|\Theta}(x \mid \theta)$ may not be available:
- $f_{\Theta|X}(\theta \mid x) = \frac{f_{\Theta}(\theta) f_{X|\Theta}(x \mid \theta)}{f_X(x)}$
- $f_X(x) = \int f_{\Theta}(\theta') f_{X|\Theta}(x \mid \theta') \, d\theta'$

Hard to implement, compute, and analyze $E[\Theta \mid X=x]$
- $E[\Theta_j \mid X=x] = \int \dots \int \theta_j f_{\Theta|X}(\theta \mid x) \, d\theta_1 \dots d\theta_n$


####**Linear Least Mean Square (LLMS) Estimation**

**LLMS Formulation**
- Unknown $\Theta$; observation $X$
- Minimize $E[(\hat{\Theta} - \Theta)^2]$
- Estimators $\Theta - g(X) \implies \hat{\Theta}_{\text{LMS}} = E[\Theta \mid X]$
- Consider estimators of $\Theta$ of the form $\hat{\Theta} = aX + b$
- Minimize $E[(\Theta - aX - b)^2]$ with respect to $a, b$
- If $E[\Theta \mid X]$ is linear in $X$, then $\hat{\Theta}_{\text{LMS}} = \hat{\Theta}_{\text{LLMS}}$

**Solution to LLMS Problem**

Minimize $E[(\Theta - aX - b)^2]$
- $\hat{\Theta}_L = E[\Theta] + \frac{\text{Cov}(\Theta, X)}{\text{Var}(X)}(X - E[X]) = E[\Theta] + \rho \frac{\sigma_{\Theta}}{\sigma_X}(X - E[X])$

For $\rho = \frac{\text{Cov}(\Theta, X)}{\sigma_{\Theta} \sigma_X} ; a = \rho \frac{\sigma_{\Theta} \sigma_X}{\sigma_X^2}$

In the solution and the error variance, only means, variances, and covariances matter
- $E[(\hat{\Theta}_L - \Theta)^2] = (1 - \rho^2) \text{Var}(\Theta)$

####**LLMS for Inferring the Parameter of a Coin**

Standard Example

- Coin with bias $\Theta$; prior $f_{\Theta}(\theta)$Fix $n$; $X = \text{number-of-heads}$Assumption$f_{\Theta}(\theta)$ is uniform in $[0,1]$
- Assumption$f_{\Theta}(\theta)$ is uniform in $[0,1]$

LMS and LLMS Estimators:
- $\hat{\Theta}_{LMS} = \frac{X + 1}{n + 2} = \hat{\Theta}_{LLMS}$

General LLMS Formula:
- $\hat{\Theta}_{LLMS} = E[\Theta] + \frac{\text{Cov}(\Theta, X)}{\text{Var}(X)}(X - E[X])$




####**LLMS with Multiple Observations**

For the model where $X_i$ represents observations of a parameter $\Theta$ with added noise:

- $X_1 = \Theta + W_1, \quad \Theta \sim (x_0, \sigma_0^2), \quad W_i \sim (0, \sigma_i^2)$
- $X_n = \Theta + W_n, \quad \Theta, W_1, \dots, W_n \text{ uncorrelated}$

Normal Distribution Case - Suppose $\Theta, W_1, \dots, W_n$ are independent normal random variables.

The Least Mean Squares (LMS) estimator is given by:
- $\hat{\theta}_{LMS} = E[\Theta \mid X_1, \dots, X_n] = \frac{\sum_{i=0}^{n} \frac{X_i}{\sigma_i^2}}{\sum_{i=0}^{n} \frac{1}{\sigma_i^2}}$

(Note: Here $X_0 = x_0$ represents the prior mean).

General Case - Suppose the distributions are general (not necessarily normal), but maintain the same means and variances as the normal example:
- All covariances remain the same.
- LLMS solution must be the same as the LMS solution in the normal case.
